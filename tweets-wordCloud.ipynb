# ツイートの前処理

```python
emojis = ''
for key in emoji.UNICODE_EMOJI.keys():
    emojis += "".join(emoji.UNICODE_EMOJI[key].keys())
    
emojis[:100]
```

```python
puncs = string.punctuation +  "◆▼★②●☆■★【】『』「」、♪"
puncs
```

```python
#半角から全角
def han2zen(txt):
    txt = jctconv.h2z(txt, kana=True, digit=False, ascii=False)
    txt = jctconv.z2h(txt, kana=False, digit=True, ascii=True)
    return txt

def remove_signs(txt):
    rm_signs = emojis + puncs
    for sign in rm_signs:
        txt = txt.replace(sign, " ")
        
    return txt

def get_nouns(txt):
    tokens = tkn.tokenize(txt)
    return " ".join([
        token.surface.strip()
        for token in tokens
        if token.part_of_speech.split(",")[0] == '名詞'
    ])

def clean_txt(txt):
    txt = han2zen(txt)
    txt = remove_signs(txt)
    txt_list = txt.lower().split()
    txt_list = [x for x in txt_list if len(x) > 1]
    txt = " ".join(list(Counter(txt_list)))
    return get_nouns(txt)
```

```python
text = []

for i in range(len(data)):
    text.append(data[i].text)
    
text_merge = ''.join(text)
#前処理して単語ごとに分ける
clean_text_merge = clean_txt(text_merge)

tokens = clean_text_merge
```

```python
#WordCloudで表示
fpath = '/System/Library/Fonts/ヒラギノ明朝 ProN.ttc'

stop_words = [u'ジャルジャル', u'後藤', u'福徳', u'さん', u'ん',u'sirup']

max_words = 150

wordcloud = WordCloud(
    font_path = fpath,
      width=1500,
      height=600, 
      background_color="white", 
      collocations = False,
      max_words=max_words,
      #max_font_size=500, 
      normalize_plurals=False,
      stopwords=set(stop_words)
)
wordcloud.generate(tokens)
```
