```python
from requests_oauthlib import OAuth1Session
import json
import pandas as pd
```

```python
import json
json_open = open('../sercret.json', 'r')
json_load = json.load(json_open)
```

```python
#èªè¨¼
CK = json_load["CK"]    # Consumer Key
CS = json_load["CS"]   # Consumer Secret
AT = json_load["AT"]    # Access Token
AS = json_load["AS"]    # Accesss Token Secert
```

```python
twitter_api = OAuth1Session(CK, CS, AT, AS)
```

```python
# å–å¾—ãƒ„ã‚¤ãƒ¼ãƒˆæ•°
count = 100

# æ¤œç´¢ãƒ¯ãƒ¼ãƒ‰
search_word = 'ã‚¸ãƒ£ãƒ«ã‚¸ãƒ£ãƒ« -filter:retweets'

# ãƒ„ã‚¤ãƒ¼ãƒˆã®æ¤œç´¢ãƒ»å–å¾—
url = 'https://api.twitter.com/1.1/search/tweets.json'
params = {'q':search_word, 'lang':'ja', 'count':count}
results = twitter_api.get(url, params=params)
results = json.loads(results.text)

# Pandasã®ãƒ‡ãƒ¼ã‚¿ãƒ•ãƒ¬ãƒ¼ãƒ ã«å…¥ã‚Œã‚‹
data_raw = pd.DataFrame(results['statuses'])
data_raw
```

```python
# ã»ã—ã„åˆ—ã ã‘æŠ½å‡º
data_new = data_raw[['created_at', 'user', 'text', 'favorite_count', 'retweet_count']]
data_new
```

```python
# ãƒ¦ãƒ¼ã‚¶ãƒ¼åã‚’è¡¨ç¤º
for result in results['statuses']:
    print(result['user']['name'])
```

```python
# ãƒ¦ãƒ¼ã‚¶ãƒ¼idã¨ãƒ¦ãƒ¼ã‚¶ãƒ¼åã‚’åˆ¥ã€…ã«å–å¾—
user_id =[]
user_name = []

for result in results['statuses']:
    user_id.append(result['user']['id'])
    user_name.append(result['user']['name'])

# data_newã«ãƒ¦ãƒ¼ã‚¶ãƒ¼idã¨ãƒ¦ãƒ¼ã‚¶ãƒ¼åã‚’è¿½åŠ 
data_new.insert(2, 'user_id', user_id)
data_new.insert(3, 'user_name', user_name)
data_new
```

```python
# ã„ã„ã­ã®å¤šã„é †ã«ä¸¦ã¹æ›¿ãˆ
data_sorted = data_new.sort_values('favorite_count', ascending=False)
data_sorted
```
